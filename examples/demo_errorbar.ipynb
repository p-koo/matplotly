## 1. Basic errorbar (connected)

```python
x = np.linspace(0, 10, 15)
y = np.sin(x)
yerr = 0.1 + 0.15 * np.abs(np.cos(x))

fig, ax = plt.subplots(figsize=(8, 4))
ax.errorbar(x, y, yerr=yerr, label='sin(x)', capsize=3, linewidth=1.5)
ax.set_title('Basic Errorbar')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend()

matplotly(fig)
```

```python
x = np.linspace(0, 10, 15)
y = np.sin(x)
yerr = 0.1 + 0.15 * np.abs(np.cos(x))

fig, ax = plt.subplots(figsize=(8, 4))
ax.errorbar(x, y, yerr=yerr, label='sin(x)', capsize=3, linewidth=1.5)
ax.set_title('Basic Errorbar')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend()

# Generated by matplotly v4

# Errorbar: sin(x)
from matplotlib.container import ErrorbarContainer as _EBC
_eb_containers = [c for c in ax.containers if isinstance(c, _EBC)]
_eb = _eb_containers[0]
if _eb[0] is not None:
    _eb[0].set_color('#1f77b4')
    _eb[0].set_linewidth(1.5)
for _cap in _eb[1]:
    _cap.set_color('#1f77b4')
    _cap.set_markersize(0)
for _bar in _eb[2]:
    _bar.set_color('#1f77b4')
    _bar.set_linewidth(0)
_segs = _eb[2][0].get_segments()
if _segs:
    _x_s = np.array([s[0][0] for s in _segs])
    _y_lo = np.array([s[0][1] for s in _segs])
    _y_hi = np.array([s[1][1] for s in _segs])
    ax.fill_between(_x_s, _y_lo, _y_hi, color='#1f77b4', alpha=0.3)

# Labels
ax.set_title('Basic Errorbar')
ax.set_xlabel('x')
ax.set_ylabel('y')

# Limits
ax.set_xlim(-0.5, 10.5)
ax.set_ylim(-1.21, 1.22)

# Fonts
ax.title.set_fontsize(12.0)
ax.title.set_fontfamily('sans-serif')
ax.xaxis.label.set_fontsize(10.0)
ax.yaxis.label.set_fontsize(10.0)
ax.tick_params(labelsize=10.0)
ax.tick_params(direction='out', length=3.5, width=0.8)

# Spines
ax.spines['top'].set_linewidth(0.8)
ax.spines['right'].set_linewidth(0.8)
ax.spines['bottom'].set_linewidth(0.8)
ax.spines['left'].set_linewidth(0.8)

# Legend
ax.legend(loc='upper right', frameon=False, fontsize=10.0, bbox_to_anchor=(1.0, 1.0))

# Layout
fig.set_size_inches(8.0, 4.0)
fig.tight_layout()
```

## 3. Multiple errorbars

## 6. Errorbars in subplots

```python
np.random.seed(11)
x = np.linspace(0, 5, 10)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# Left: connected errorbar
y1 = np.exp(-0.3 * x) * np.cos(x)
yerr1 = 0.05 + 0.1 * np.abs(np.sin(x))
ax1.errorbar(x, y1, yerr=yerr1, label='decay', capsize=3, linewidth=1.5)
ax1.set_title('Decaying cosine')
ax1.set_xlabel('Time')
ax1.set_ylabel('Amplitude')
ax1.legend()

# Right: scatter errorbar
y2 = 0.5 * x + np.random.randn(len(x)) * 0.3
yerr2 = 0.2 + 0.2 * np.random.rand(len(x))
ax2.errorbar(x, y2, yerr=yerr2, fmt='D', linestyle='none',
             label='linear trend', capsize=4, color='#e67e22')
ax2.set_title('Linear trend')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.legend()

fig.tight_layout()
matplotly(fig)
```
