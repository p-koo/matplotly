## Test 7 — Styled box + violin overlay (manual)

```python
rng = np.random.default_rng(2024)

fig, ax = plt.subplots(figsize=(7, 5))

# Create violin + boxplot overlay manually
data = [rng.normal(0, 1, 300),
        rng.gamma(2, 1, 300),
        rng.standard_t(5, 300)]

vp = ax.violinplot(data, positions=[1, 2, 3], showextrema=False)
for body in vp['bodies']:
    body.set_alpha(0.3)

bp = ax.boxplot(data, positions=[1, 2, 3], widths=0.15,
                patch_artist=True, showfliers=False)
for patch in bp['boxes']:
    patch.set_facecolor('white')
    patch.set_edgecolor('black')

ax.set_xticks([1, 2, 3])
ax.set_xticklabels(['Normal', 'Gamma', 'Student-t'])
ax.set_title('Distribution Comparison')
ax.set_ylabel('Value')

matplotly(fig)
```

```python
rng = np.random.default_rng(777)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

# Left: box plot
data1 = [rng.normal(100, 15, 80) for _ in range(4)]
# ax1.boxplot(data1, tick_labels=['A', 'B', 'C', 'D'], patch_artist=True)
ax1.set_title('Box Plot')
ax1.set_ylabel('Value')

# Right: violin plot
data2 = [rng.normal(100, 15, 80) for _ in range(4)]
# ax2.violinplot(data2, positions=[1, 2, 3, 4])
ax2.set_xticks([1, 2, 3, 4])
ax2.set_xticklabels(['W', 'X', 'Y', 'Z'])
ax2.set_title('Violin Plot')
ax2.set_ylabel('Value')

fig.suptitle('Side-by-Side Comparison', fontsize=14)
fig.tight_layout()

# Generated by matplotly v4
axes = fig.get_axes()

# Distribution plot

# (4 boxes)
_data = [data1[i] for i in [0, 1, 2, 3]]
_vp = axes[0].violinplot(_data,
    positions=[1.0, 2.0, 3.0, 4.0],
    widths=0.9, vert=True,
    showmeans=False, showmedians=False, showextrema=False)
for _body in _vp['bodies']:
    _body.set_facecolor('#1f77b4')
    _body.set_edgecolor('#000000')
    _body.set_alpha(0.3)
_bp = axes[0].boxplot(_data,
    positions=[1.0, 2.0, 3.0, 4.0],
    vert=True,
    patch_artist=True,
    widths=0.18,
    showfliers=False,
    boxprops=dict(facecolor='#1f77b4', edgecolor='#000000'),
    medianprops=dict(color='#ff7f0e'))
_bp['boxes'][0].set_label('(4 boxes)')
_rng_j = np.random.RandomState(42)
for _i, (_d, _pos) in enumerate(zip(_data, [1.0, 2.0, 3.0, 4.0])):
    _jitter = _rng_j.uniform(-0.2, 0.2, len(_d))
    axes[0].scatter(_pos + _jitter, _d, s=9.0, c='#1f77b4', alpha=0.5, marker='o', zorder=3)

axes[0].set_xticks([1.0, 2.0, 3.0, 4.0])
axes[0].set_xticklabels(['A', 'B', 'C', 'D'])

# Labels
axes[0].set_title('Box Plot')
axes[0].set_ylabel('Value')

# Limits
axes[0].set_xlim(0.4, 4.6)
axes[0].set_ylim(57.77, 143.19)

# Fonts
axes[0].title.set_fontsize(10.0)
axes[0].title.set_fontfamily('Arial')
axes[0].xaxis.label.set_fontsize(10.0)
axes[0].yaxis.label.set_fontsize(10.0)
axes[0].tick_params(labelsize=10.0)
axes[0].tick_params(direction='out', length=3.5, width=0.8)
axes[0].xaxis.set_major_locator(matplotlib.ticker.MultipleLocator(1.0))
axes[0].yaxis.set_major_locator(matplotlib.ticker.MultipleLocator(20.0))

# Spines
axes[0].spines['top'].set_visible(False)
axes[0].spines['top'].set_linewidth(0.8)
axes[0].spines['right'].set_visible(False)
axes[0].spines['right'].set_linewidth(0.8)
axes[0].spines['bottom'].set_linewidth(0.8)
axes[0].spines['left'].set_linewidth(0.8)

# Legend removed
if axes[0].get_legend() is not None:
    axes[0].get_legend().remove()

# Distribution plot

# (4 violins)
_data = [data1[i] for i in [0, 1, 2, 3]]
_vp = axes[1].violinplot(_data,
    positions=[1.0007, 2.0002, 3.0004, 4.0005],
    widths=0.9, vert=True,
    showmeans=False, showmedians=False, showextrema=False)
for _body in _vp['bodies']:
    _body.set_facecolor('#ff7f0e')
    _body.set_edgecolor('#1f77b4')
    _body.set_alpha(0.3)
_bp = axes[1].boxplot(_data,
    positions=[1.0007, 2.0002, 3.0004, 4.0005],
    vert=True,
    patch_artist=True,
    widths=0.18,
    showfliers=False,
    boxprops=dict(facecolor='#ff7f0e', edgecolor='#1f77b4'),
    medianprops=dict(color='#1f77b4'))
_bp['boxes'][0].set_label('(4 violins)')
_rng_j = np.random.RandomState(42)
for _i, (_d, _pos) in enumerate(zip(_data, [1.0007, 2.0002, 3.0004, 4.0005])):
    _jitter = _rng_j.uniform(-0.2, 0.2, len(_d))
    axes[1].scatter(_pos + _jitter, _d, s=9.0, c='#ff7f0e', alpha=0.5, marker='o', zorder=3)

axes[1].set_xticks([1.0007, 2.0002, 3.0004, 4.0005])
axes[1].set_xticklabels(['W', 'X', 'Y', 'Z'])

# Labels
axes[1].set_title('Violin Plot')
axes[1].set_ylabel('Value')

# Limits
axes[1].set_xlim(0.4, 4.6)
axes[1].set_ylim(49.11, 143.8)

# Fonts
axes[1].title.set_fontsize(10.0)
axes[1].title.set_fontfamily('Arial')
axes[1].xaxis.label.set_fontsize(10.0)
axes[1].yaxis.label.set_fontsize(10.0)
axes[1].tick_params(labelsize=10.0)
axes[1].tick_params(direction='out', length=3.5, width=0.8)

# Spines
axes[1].spines['top'].set_visible(False)
axes[1].spines['top'].set_linewidth(0.8)
axes[1].spines['right'].set_visible(False)
axes[1].spines['right'].set_linewidth(0.8)
axes[1].spines['bottom'].set_linewidth(0.8)
axes[1].spines['left'].set_linewidth(0.8)

# Legend removed
if axes[1].get_legend() is not None:
    axes[1].get_legend().remove()

# Layout
fig.set_size_inches(7.5, 3.7)
fig.tight_layout()
fig.subplots_adjust(hspace=0.2, wspace=0.18)
```
