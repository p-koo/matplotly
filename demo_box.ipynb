```python
rng = np.random.default_rng(42)

fig, ax = plt.subplots(figsize=(7, 5))

data = [rng.normal(0, 1, 100),
        rng.normal(2, 1.5, 100),
        rng.normal(1, 0.8, 100)]
# ax.boxplot(data, tick_labels=['Control', 'Treatment A', 'Treatment B'])

ax.set_title('Drug Trial Results')
ax.set_ylabel('Response')

# Generated by matplotly v4

# Distribution plot

# (3 boxes)
_data = [data[i] for i in [0, 1, 2]]
_vp = ax.violinplot(_data,
    positions=[1.0, 2.0, 3.0],
    widths=0.9, vert=True,
    showmeans=False, showmedians=False, showextrema=False)
for _body in _vp['bodies']:
    _body.set_facecolor('#1f77b4')
    _body.set_edgecolor('#000000')
    _body.set_alpha(0.3)
_bp = ax.boxplot(_data,
    positions=[1.0, 2.0, 3.0],
    vert=True,
    patch_artist=True,
    widths=0.18,
    showfliers=False,
    boxprops=dict(facecolor='#1f77b4', edgecolor='#000000'),
    medianprops=dict(color='#ff7f0e'))
_bp['boxes'][0].set_label('(3 boxes)')
_rng_j = np.random.RandomState(42)
for _i, (_d, _pos) in enumerate(zip(_data, [1.0, 2.0, 3.0])):
    _jitter = _rng_j.uniform(-0.2, 0.2, len(_d))
    ax.scatter(_pos + _jitter, _d, s=9.0, c='#1f77b4', alpha=0.5, marker='o', zorder=3)

ax.set_xticks([1.0, 2.0, 3.0])
ax.set_xticklabels(['Control', 'Treatment A', 'Treatment B'])

# Labels
ax.set_title('Drug Trial Results')
ax.set_ylabel('Response')

# Limits
ax.set_xlim(0.4, 3.6)
ax.set_ylim(-2.34, 6.79)

# Fonts
ax.title.set_fontsize(10.0)
ax.title.set_fontfamily('Arial')
ax.xaxis.label.set_fontsize(10.0)
ax.yaxis.label.set_fontsize(10.0)
ax.tick_params(labelsize=10.0)
ax.tick_params(direction='out', length=3.5, width=0.8)

# Spines
ax.spines['top'].set_visible(False)
ax.spines['top'].set_linewidth(0.8)
ax.spines['right'].set_visible(False)
ax.spines['right'].set_linewidth(0.8)
ax.spines['bottom'].set_linewidth(0.8)
ax.spines['left'].set_linewidth(0.8)

# Legend
ax.legend(loc='upper right', frameon=False, fontsize=10.0, bbox_to_anchor=(1.0, 1.0))

# Layout
fig.set_size_inches(3.4, 3.4)
fig.tight_layout()
```

## Test 3 — Violin plot

```python
rng = np.random.default_rng(7)

fig, ax = plt.subplots(figsize=(7, 5))

# Bimodal distribution to show violin shape advantage
d1 = np.concatenate([rng.normal(-1, 0.5, 150), rng.normal(1.5, 0.4, 100)])
d2 = rng.normal(0, 1, 250)
d3 = rng.exponential(1, 250)

vp = ax.violinplot([d1, d2, d3], positions=[1, 2, 3],
                   showmeans=True, showmedians=True)

ax.set_xticks([1, 2, 3])
ax.set_xticklabels(['Bimodal', 'Normal', 'Exponential'])
ax.set_title('Distribution Shapes')
ax.set_ylabel('Value')

matplotly(fig)
```

## Test 5 — Many boxes with outliers

```python
rng = np.random.default_rng(55)

fig, ax = plt.subplots(figsize=(9, 5))

# Seven groups with different spreads and outliers
labels = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
data = []
for i, day in enumerate(labels):
    base = rng.normal(50 + i * 3, 8 + i, 60)
    outliers = rng.uniform(90, 120, rng.integers(1, 5))
    data.append(np.concatenate([base, outliers]))

ax.boxplot(data, tick_labels=labels, notch=True)

ax.set_title('Daily Response Times (ms)')
ax.set_xlabel('Day of Week')
ax.set_ylabel('Response Time (ms)')

matplotly(fig)
```
