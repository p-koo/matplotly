## Test 1 — Pure scatter (two groups)

```python
rng = np.random.default_rng(42)

fig, ax = plt.subplots(figsize=(7, 5))

# Group A: clustered around (2, 3)
n_a = 60
x_a = rng.normal(2, 0.8, n_a)
y_a = rng.normal(3, 1.0, n_a)
ax.scatter(x_a, y_a, s=40, alpha=0.7, label='Group A')

# Group B: clustered around (5, 6)
n_b = 50
x_b = rng.normal(5, 1.2, n_b)
y_b = rng.normal(6, 0.9, n_b)
ax.scatter(x_b, y_b, s=55, alpha=0.7, label='Group B')

ax.set_title('Two Clusters')
ax.set_xlabel('Feature 1')
ax.set_ylabel('Feature 2')
ax.legend()

matplotly(fig)
```

## Test 2 — Mixed line + scatter

```python
rng = np.random.default_rng(7)

x = np.linspace(0, 10, 200)
y_true = 2.5 * np.sin(x) + 0.5 * x

# Noisy observations
n_obs = 40
x_obs = rng.uniform(0, 10, n_obs)
y_obs = 2.5 * np.sin(x_obs) + 0.5 * x_obs + rng.normal(0, 1.0, n_obs)

fig, ax = plt.subplots(figsize=(8, 4.5))
ax.plot(x, y_true, linewidth=2, label='True signal')
ax.scatter(x_obs, y_obs, s=35, zorder=5, label='Observations')

ax.set_title('Signal vs. Noisy Observations')
ax.set_xlabel('Time (s)')
ax.set_ylabel('Amplitude')
ax.legend()

matplotly(fig)
```

```python
rng = np.random.default_rng(7)

x = np.linspace(0, 10, 200)
y_true = 2.5 * np.sin(x) + 0.5 * x

# Noisy observations
n_obs = 40
x_obs = rng.uniform(0, 10, n_obs)
y_obs = 2.5 * np.sin(x_obs) + 0.5 * x_obs + rng.normal(0, 1.0, n_obs)

fig, ax = plt.subplots(figsize=(8, 4.5))
ax.plot(x, y_true, linewidth=2, label='True signal')
ax.scatter(x_obs, y_obs, s=35, zorder=5, label='Observations')

ax.set_title('Signal vs. Noisy Observations')
ax.set_xlabel('Time (s)')
ax.set_ylabel('Amplitude')
ax.legend()

# Generated by matplotly

# True signal
ax.lines[0].set_color('#d62728')
ax.lines[0].set_linewidth(2.0)
ax.lines[0].set_linestyle('--')
ax.collections[0].set_facecolor('#1f77b4')

# Labels
ax.set_title('Signal vs. Noisy Observations')
ax.set_xlabel('Time (s)')
ax.set_ylabel('Amplitude')

# Limits
ax.set_xlim(-0.5, 10.5)
ax.set_ylim(-2.09, 7.76)

# Fonts
ax.title.set_fontsize(12.0)
ax.title.set_fontfamily('sans-serif')
ax.xaxis.label.set_fontsize(10.0)
ax.yaxis.label.set_fontsize(10.0)
ax.tick_params(labelsize=10.0)
ax.tick_params(direction='out', length=3.5, width=0.8)

# Spines
ax.spines['top'].set_visible(False)
ax.spines['top'].set_linewidth(0.8)
ax.spines['right'].set_visible(False)
ax.spines['right'].set_linewidth(0.8)
ax.spines['bottom'].set_linewidth(0.8)
ax.spines['left'].set_linewidth(0.8)

# Legend
handles, labels = ax.get_legend_handles_labels()
for h, lbl in zip(handles, ['True signal', 'Observations']):
    h.set_label(lbl)
ax.legend(loc='best', frameon=False, fontsize=10.0)

# Layout
fig.set_size_inches(5.0, 3.6)
fig.tight_layout()
```

## Test 3 — Three scatter groups with varying size

```python
rng = np.random.default_rng(123)

fig, ax = plt.subplots(figsize=(7, 5))

for i, (label, center, size) in enumerate([
    ('Control', (1, 4), 30),
    ('Treatment A', (4, 7), 50),
    ('Treatment B', (6, 2), 70),
]):
    n = 45
    x = rng.normal(center[0], 1.0, n)
    y = rng.normal(center[1], 1.2, n)
    ax.scatter(x, y, s=size, alpha=0.65, edgecolors='white',
              linewidths=0.8, label=label)

ax.set_title('Experiment Results')
ax.set_xlabel('Dose (mg)')
ax.set_ylabel('Response')
ax.legend()

matplotly(fig)
```
