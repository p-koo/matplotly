```python
import sys
sys.path.insert(0, '.')

import numpy as np
import matplotlib.pyplot as plt
from matplotly import matplotly                                                                                       
```

```python
# Fake data: damped sine wave
x = np.linspace(0, 4 * np.pi, 200)
y = np.exp(-0.15 * x) * np.sin(x)

x2 = np.linspace(0, 4 * np.pi, 200)
y2 = np.exp(-0.15 * x) * np.sin(x+1.2)

x3 = np.linspace(0, 4 * np.pi, 200)
y3 = np.exp(-0.15 * x) * np.sin(x+2)

fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(x, y, label='damped sine', linewidth=1.5)
ax.plot(x2, y2, label='damped sine', linewidth=1.5)
ax.plot(x3, y3, label='damped sine', linewidth=1.5)
ax.set_title('Damped Oscillation')
ax.set_xlabel('Time (s)')
ax.set_ylabel('Amplitude')


matplotly(fig)
```
