```python
fig, ax = plt.subplots(figsize=(7, 5))

ax.bar(['Cat', 'Dog', 'Bird'], [10, 20, 30], label='Counts')

ax.set_title('Pet Popularity')
ax.set_xlabel('Animal')
ax.set_ylabel('Count')
ax.legend()

# Generated by matplotly v4

# Bars — remove originals and recreate
from matplotlib.container import BarContainer as _BC
for _c in list(ax.containers):
    if isinstance(_c, _BC):
        for _p in _c: _p.remove()
ax.containers[:] = [_c for _c in ax.containers if not isinstance(_c, _BC)]
_x = np.array([0.0, 1.0, 2.0])
ax.bar(_x, [10.0, 20.0, 30.0], width=0.8, color='#1f77b4', edgecolor='#000000', linewidth=1.0, alpha=1.0, label='Counts')
ax.set_xticks(_x)
ax.set_xticklabels(['Cat', 'Dog', 'Bird'])
from matplotlib.ticker import AutoLocator, ScalarFormatter
ax.yaxis.set_major_locator(AutoLocator())
ax.yaxis.set_major_formatter(ScalarFormatter())

# Labels
ax.set_title('Pet Popularity')
ax.set_xlabel('Animal')
ax.set_ylabel('Count')

# Limits
ax.set_xlim(-0.54, 2.54)
ax.set_ylim(0.0, 31.5)

# Fonts
ax.title.set_fontsize(12.0)
ax.title.set_fontfamily('sans-serif')
ax.xaxis.label.set_fontsize(10.0)
ax.yaxis.label.set_fontsize(10.0)
ax.tick_params(labelsize=10.0)
ax.tick_params(direction='out', length=3.5, width=0.8)

# Spines
ax.spines['top'].set_linewidth(0.8)
ax.spines['right'].set_linewidth(0.8)
ax.spines['bottom'].set_linewidth(0.8)
ax.spines['left'].set_linewidth(0.8)

# Legend removed
if ax.get_legend() is not None:
    ax.get_legend().remove()

# Layout
fig.set_size_inches(7.0, 5.0)
fig.tight_layout()
```

## Test 2 — Grouped bar chart

```python
fig, ax = plt.subplots(figsize=(7, 5))

x = np.arange(4)
ax.bar(x - 0.2, [20, 35, 30, 25], 0.4, label='2023')
ax.bar(x + 0.2, [25, 32, 34, 28], 0.4, label='2024')

ax.set_xticks(x)
ax.set_xticklabels(['Q1', 'Q2', 'Q3', 'Q4'])
ax.set_title('Quarterly Revenue')
ax.set_xlabel('Quarter')
ax.set_ylabel('Revenue ($M)')
ax.legend()

# Generated by matplotly v4

# 2023
for _p in ax.containers[0]:
    _p.set_facecolor('#1f77b4')
    _p.set_edgecolor('#000000')
for _p, _x in zip(ax.containers[0], [-0.15, 0.85, 1.85, 2.85]):
    _p.set_width(0.1)
    _p.set_x(_x)

# 2024
for _p in ax.containers[1]:
    _p.set_facecolor('#ff7f0e')
    _p.set_edgecolor('#000000')
for _p, _x in zip(ax.containers[1], [0.05, 1.05, 2.05, 3.05]):
    _p.set_width(0.1)
    _p.set_x(_x)
ax.set_xticks([0.0, 1.0, 2.0, 3.0])
ax.set_xticklabels(['Q1', 'Q2', 'Q3', 'Q4'])

# Labels
ax.set_title('Quarterly Revenue')
ax.set_xlabel('Quarter')
ax.set_ylabel('Revenue ($M)')

# Limits
ax.set_xlim(-0.32, 3.31)
ax.set_ylim(0.0, 36.75)

# Fonts
ax.title.set_fontsize(10.0)
ax.title.set_fontfamily('Arial')
ax.xaxis.label.set_fontsize(10.0)
ax.yaxis.label.set_fontsize(10.0)
ax.tick_params(labelsize=10.0)
ax.tick_params(direction='out', length=3.5, width=0.8)

# Spines
ax.spines['top'].set_visible(False)
ax.spines['top'].set_linewidth(0.8)
ax.spines['right'].set_visible(False)
ax.spines['right'].set_linewidth(0.8)
ax.spines['bottom'].set_linewidth(0.8)
ax.spines['left'].set_linewidth(0.8)

# Legend
ax.legend(loc='upper right', frameon=False, fontsize=10.0, bbox_to_anchor=(1.0, 1.0))

# Layout
fig.set_size_inches(5.3, 3.7)
fig.tight_layout()
```

## Test 4 — Three-group bar chart

```python
fig, ax = plt.subplots(figsize=(8, 5))

categories = ['Math', 'Science', 'English', 'History', 'Art']
x = np.arange(len(categories))
w = 0.25

ax.bar(x - w, [88, 92, 75, 80, 95], w, label='Alice')
ax.bar(x,     [76, 85, 90, 88, 70], w, label='Bob')
ax.bar(x + w, [95, 78, 82, 75, 88], w, label='Carol')

ax.set_xticks(x)
ax.set_xticklabels(categories)
ax.set_title('Student Scores by Subject')
ax.set_ylabel('Score')
ax.set_ylim(0, 105)
ax.legend()

matplotly(fig)
```

```python
fig, ax = plt.subplots(figsize=(8, 5))

categories = ['Math', 'Science', 'English', 'History', 'Art']
x = np.arange(len(categories))
w = 0.25

ax.bar(x - w, [88, 92, 75, 80, 95], w, label='Alice')
ax.bar(x,     [76, 85, 90, 88, 70], w, label='Bob')
ax.bar(x + w, [95, 78, 82, 75, 88], w, label='Carol')

ax.set_xticks(x)
ax.set_xticklabels(categories)
ax.set_title('Student Scores by Subject')
ax.set_ylabel('Score')
ax.set_ylim(0, 105)
ax.legend()

# Generated by matplotly v4

# Alice
for _p in ax.containers[0]:
    _p.set_facecolor('#1b9e77')
    _p.set_edgecolor('#000000')
for _p, _x in zip(ax.containers[0], [-0.2, 0.8, 1.8, 2.8, 3.8]):
    _p.set_width(0.1)
    _p.set_x(_x)

# Bob
for _p in ax.containers[1]:
    _p.set_facecolor('#d95f02')
    _p.set_edgecolor('#000000')
for _p, _x in zip(ax.containers[1], [-0.05, 0.95, 1.95, 2.95, 3.95]):
    _p.set_width(0.1)
    _p.set_x(_x)

# Carol
for _p in ax.containers[2]:
    _p.set_facecolor('#7570b3')
    _p.set_edgecolor('#000000')
for _p, _x in zip(ax.containers[2], [0.1, 1.1, 2.1, 3.1, 4.1]):
    _p.set_width(0.1)
    _p.set_x(_x)
ax.set_xticks([0.0, 1.0, 2.0, 3.0, 4.0])
ax.set_xticklabels(['Math', 'Science', 'English', 'History', 'Art'])

# Labels
ax.set_title('Student Scores by Subject')
ax.set_ylabel('Score')

# Limits
ax.set_xlim(-0.42, 4.42)
ax.set_ylim(0.0, 105.0)

# Fonts
ax.title.set_fontsize(10.0)
ax.title.set_fontfamily('Arial')
ax.xaxis.label.set_fontsize(10.0)
ax.yaxis.label.set_fontsize(10.0)
ax.tick_params(labelsize=10.0)
ax.tick_params(direction='out', length=3.5, width=0.8)

# Spines
ax.spines['top'].set_visible(False)
ax.spines['top'].set_linewidth(0.8)
ax.spines['right'].set_visible(False)
ax.spines['right'].set_linewidth(0.8)
ax.spines['bottom'].set_linewidth(0.8)
ax.spines['left'].set_linewidth(0.8)

# Legend
ax.legend(loc='upper right', frameon=False, fontsize=10.0, ncol=4, bbox_to_anchor=(0.8, 1.02))

# Layout
fig.set_size_inches(6.0, 3.7)
fig.tight_layout()
```

## Test 5 — Pre-styled bars (hatches, alpha, edge colors)

```python
fig, ax = plt.subplots(figsize=(7, 5))

x = np.arange(3)
ax.bar(x - 0.2, [40, 55, 30], 0.4,
       label='Actual', color='#2196F3', edgecolor='#0D47A1',
       linewidth=1.5, hatch='//', alpha=0.8)
ax.bar(x + 0.2, [50, 45, 35], 0.4,
       label='Target', color='#FF9800', edgecolor='#E65100',
       linewidth=1.5, hatch='xx', alpha=0.8)

ax.set_xticks(x)
ax.set_xticklabels(['Jan', 'Feb', 'Mar'])
ax.set_title('Actual vs Target Sales')
ax.set_ylabel('Units')
ax.legend()

matplotly(fig)
```
